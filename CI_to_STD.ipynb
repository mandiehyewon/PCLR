```python
'''
Last Edited: 07/26/23
'''

import scipy.stats as st

def std_from_confidence_interval(mean, ci_lower, ci_upper, confidence_level=0.95):
    # Calculate the width of the confidence interval
    ci_width = ci_upper - ci_lower

    # Get the z-score corresponding to the desired confidence level
    z_score = st.norm.ppf((1 + confidence_level) / 2)

    # Calculate the standard deviation from the confidence interval
    std_estimate = ci_width / (2 * z_score)

    return std_estimate
```
